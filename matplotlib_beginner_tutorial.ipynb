### 5.1 åŸºæœ¬ã®æ•£å¸ƒå›³

### 5.2 ç‚¹ã®å¤§ãã•ã¨è‰²

````python
np.random.seed(42)
x = np.random.randn(50)
y = np.random.randn(50)
colors = np.random.rand(50)      # è‰²ï¼ˆ0ã€œ1ã®å€¤ï¼‰
sizes = np.random.rand(50) * 200  # ç‚¹ã®ã‚µã‚¤ã‚º

plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
plt.colorbar(label='è‰²ã®å€¤')
plt.title("å¤§ãã•ã¨è‰²ã‚’å¤‰ãˆãŸæ•£å¸ƒå›³")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

### 5.3 ã‚°ãƒ«ãƒ¼ãƒ—åˆ¥ã®æ•£å¸ƒå›³

````python
np.random.seed(42)

# ã‚°ãƒ«ãƒ¼ãƒ—A
x_a = np.random.randn(30) + 2
y_a = np.random.randn(30) + 2

# ã‚°ãƒ«ãƒ¼ãƒ—B
x_b = np.random.randn(30) - 1
y_b = np.random.randn(30) - 1

plt.scatter(x_a, y_a, c='red', label='ã‚°ãƒ«ãƒ¼ãƒ—A', alpha=0.7)
plt.scatter(x_b, y_b, c='blue', label='ã‚°ãƒ«ãƒ¼ãƒ—B', alpha=0.7)
plt.title("ã‚°ãƒ«ãƒ¼ãƒ—åˆ¥ã®æ•£å¸ƒå›³")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
````
