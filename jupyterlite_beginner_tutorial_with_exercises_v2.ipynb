### 第13章 練習問題

1. 上記のデータフレームに `bmi` 列を追加しなさい。ただし、簡略化のため「体重(kg) ÷ (身長(m))^2」で計算するものとする。
2. `bmi` の平均値を計算して表示しなさい。
3. 身長を x 軸、BMI を y 軸とする散布図を描きなさい。

````python
# 第13章 模範解答例

import pandas as pd
import matplotlib.pyplot as plt

data = {
    "height": [150, 160, 170, 180, 190],
    "weight": [50, 55, 63, 80, 90]
}

df = pd.DataFrame(data)

# 1. BMI 列の追加
df["height_m"] = df["height"] / 100  # cm -> m
df["bmi"] = df["weight"] / (df["height_m"] ** 2)

print(df)

# 2. BMI の平均値
print("BMIの平均値:", df["bmi"].mean())

# 3. 散布図（身長 vs BMI）
plt.scatter(df["height"], df["bmi"])
plt.xlabel("Height (cm)")
plt.ylabel("BMI")
plt.title("Height vs BMI")
plt.show()
````

# 第14章：相関係数を求める

Pandas では、`DataFrame.corr()` を使って数値列同士の相関係数を求めることができます。

````python
import pandas as pd
import numpy as np

np.random.seed(0)

df = pd.DataFrame({
    "x": np.random.randn(100),
    "y": np.random.randn(100) * 0.5 + 0.2
})

print("平均値")
print(df.mean())

print("\n相関係数")
print(df.corr())
````

### 第14章 練習問題

1. 上記のコードを実行して、`x` と `y` の相関係数がどの程度の値か確認しなさい。相関の強さについて、自分の言葉で簡単にコメントを書きなさい。
2. 新しく列 `z` を追加し、`z = x + y` と定義したとき、`x` と `z` の相関係数は `x` と `y` と比べてどうなりそうか、予想を書きなさい。
3. 実際に `z` を追加して `df.corr()` を再計算し、`x` と `z` の相関係数を確認しなさい。予想と比較してコメントしなさい。

````python
# 第14章 模範解答例

import pandas as pd
import numpy as np

np.random.seed(0)

df = pd.DataFrame({
    "x": np.random.randn(100),
    "y": np.random.randn(100) * 0.5 + 0.2
})

# 1. x と y の相関係数を確認
print("相関係数（x と y）")
print(df.corr())

# 2. 予想（コメント例）
# z = x + y と定義すると、z は x を含んでいるため、x と z の相関は
# x と y よりも強くなると予想できる。

# 3. 実際に z を追加して相関を再計算
df["z"] = df["x"] + df["y"]
print("\n相関係数（x, y, z）")
print(df.corr())

# コメント例：
# 実際に計算すると、x と z の相関係数は 0.9 前後など高い値になり、
# x と y の相関よりも強い正の相関が見られる。
````

# 第18章：micropip と piplite の違いと対応パッケージ一覧

JupyterLite（Pyodide）では、通常の `pip install` の代わりに **micropip** や **piplite** を使ってパッケージを読み込みます。  
ここでは、その違いと「どのような場面でどちらを使うか」を整理します。

## 18.1 micropip と piplite の役割の違い

次の表は、micropip と piplite の主な違いをまとめたものです。

| 項目 | micropip | piplite |
|------|----------|---------|
| 主な役割 | PyPI など外部リポジトリから **純 Python パッケージ** をインストール | JupyterLite サイト側にあらかじめ用意された **wheel（.whl）をロード** |
| 提供元 | Pyodide 公式 | JupyterLite 側の仕組み |
| インストール元 | インターネット上の PyPI | JupyterLite サーバー上の `/pypi/` など |
| 典型的な用途 | 軽量なユーティリティ・解析補助ライブラリを追加したいとき | NumPy や Pandas、自作ライブラリなどを **「授業用サイト側で配布」** したいとき |
| ネット接続 | 原則として必要 | wheel が JupyterLite と一緒に配布されていればオフラインでも可 |
| C拡張付きパッケージ | 基本的に不可（純 Python 前提） | 事前ビルド済みの wheel を置けば利用可能な場合がある |


## 18.2 micropip でインストールできるもの（概念的な一覧）

**micropip でインストールできる代表的なパターン（例）**

| 分類 | micropip で扱いやすいものの例 |
|------|------------------------------|
| 純 Python のユーティリティ | 文字列処理ライブラリ、小さめのツールキットなど |
| データ処理系（軽量） | 追加の前処理用ツール、統計補助ライブラリ（C拡張なし） |
| Web/API クライアント | HTTP クライアントなど、一部の純 Python 実装 |
| 自作パッケージ | 自分で作成し PyPI に公開した純 Python パッケージ |

※ 具体的にインストール可能かどうかは、そのパッケージが **Pyodide 上で動く純 Python かどうか** に依存します。

## 18.3 piplite でインストールできるもの（概念的な一覧）

**piplite でインストールできる代表的なパターン（例）**  
（前提：JupyterLite サイト運営者が `.whl` ファイルを用意している）

| 分類 | piplite で扱いやすいものの例 |
|------|------------------------------|
| 科学技術計算ライブラリ | NumPy, Pandas, Matplotlib など、Pyodide バンドルまたは事前ビルド済み wheel |
| 統計・機械学習ライブラリ | SciPy, scikit-learn など（Pyodide が対応している範囲） |
| 授業用・研究用ライブラリ | 教員が自作した解析用関数群を wheel 化したもの |
| オフライン授業用パッケージ | 学内ネットワーク・USB などで配布する wheel 一式 |

※ piplite は「JupyterLite サイトに置いてある wheel を読み込むための仕組み」であり、**単体で PyPI にアクセスすることはありません**。

## 18.4 micropip と piplite の基本的な使い方（チュートリアル）

### micropip の例：純 Python パッケージのインストール

```python
import micropip

# 例：純 Python のパッケージをインストールしたいとき
await micropip.install("example-package")

import example_package  # インストールしたパッケージを利用
```

### piplite の例：サイト側で用意された wheel をロード

```python
import piplite

# 例：JupyterLite サイト側が numpy と pandas の wheel を用意している場合
await piplite.install(["numpy", "pandas"])

import numpy as np
import pandas as pd

print(np.arange(5))
print(pd.DataFrame({"a": [1, 2, 3]}))
```

どちらを使うべきかの目安：

- 「**PyPI から取得したい軽量パッケージ**」 → micropip
- 「**授業サイトにあらかじめ置いたパッケージや、NumPy/Pandas などを確実に使わせたい**」 → piplite
